```python
# ctypes is a foreign function library for Python. It provides C compatible data types
```

```python
class New_List():
    def __init__(self):
        self.size = 1
        self.items = 0
        # create a ctype array whose size is equal to self.size
        self.a = self.__createarray(self.size)
    def __createarray(self,capacity):
        # creates a ctype array(static and referential) with size capacity
        return (capacity*ctypes.py_object)()
    def __len__(self):
        return self.items
    def append(self,n):
        if self.items == self.size:
            # resize
            self.__resize(self.size*2)  
        self.a[self.items] = n
        self.items += 1
    def __resize(self,new_capacity):
        b = self.__createarray(new_capacity)
        self.size = new_capacity
        # copy content of a to b
        for i in range(self.items):
            b[i] = self.a[i]
        # reassign
        self.a = b
    def __str__(self):
        result = ''
        for i in range(self.items):
            result = result + str(self.a[i]) + ','
        return '[' + result[:-1] + ']'
    def __getitem__(self,index):
        if 0 < index < self.items:
            return self.a[index]
        else:
            return 'Index Error: Index Out Of Range'
    def pop(self):
        if self.items == 0:
            return 'Empty List'
        print(self.a[self.items - 1])
        self.items = self.items - 1
        
    def clear(self):
        self.items = 0
        self.size = 1
    def find(self,n):
        for i in range(self.items):
            if self.a[i] == n:
                return i
        return 'Value Error - Not found in the list'
    def insert(self,pos,n):
        if self.items == self.size:
            self.__resize(self.size*2)
        for i in range(self.items,pos,-1):
            self.a[i] = self.a[i-1]
        self.a[pos] = n
        self.items = self.items + 1
    def __delitem__(self,pos):
        for i in range(pos,self.items -1):
            self.a[i] = self.a[i+1]
        self.items = self.items -1
    def remove(self,n):
        pos = self.find(n)
        
        if type(pos) == int:
            #delete
            self.__delitem__(pos)
        else:
            return pos
```
